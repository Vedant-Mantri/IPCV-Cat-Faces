```python
import cv2
import matplotlib.pyplot as plt

# using a test cat image
image = cv2.imread('1.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(histogram)
plt.title('Histogram of test image')
plt.show()

#1st image
image1 = cv2.imread('2.jpg')
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
histogram1 = cv2.calcHist([gray_image1], [0], None, [256], [0, 256])

#2nd image
image2 = cv2.imread('3.jpg')
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
histogram2 = cv2.calcHist([gray_image2], [0], None, [256], [0, 256])

c1 = 0 
c2 = 0

# Euclidean Distance between 1st cat image and test image
i = 0
while i<len(histogram) and i<len(histogram1):
    c1+=(histogram[i]-histogram1[i])**2
    i+= 1
c1 = c1**(1 / 2)


# Euclidean Distance between 2nd cat image and test image
j = 0
while j<len(histogram) and j<len(histogram2):
    c2+=(histogram[j]-histogram2[j])**2
    j+= 1
c2 = c2**(1 / 2)

if(c1 < c2):
    print("1st Image is more similar to the test image as compared to 2nd Image")
    plt.plot(histogram1)
    plt.title('Histogram of 1st image')
    plt.show()
else:
    print("2nd Image is more similar to the test image as compared to 1st Image")
    plt.plot(histogram2)
    plt.title('Histogram of 2nd image')
    plt.show()
    
```
